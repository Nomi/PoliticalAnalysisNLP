## Democratic Party Platform 2016

```python
url_dem_platform_2016 = 'https://www.presidency.ucsb.edu/documents/2016-democratic-party-platform'
url_debate_las_vegas_2016 = 'https://www.presidency.ucsb.edu/documents/presidential-debate-the-university-nevada-las-vegas'
url_debate_st_louis_2016 = 'https://www.presidency.ucsb.edu/documents/presidential-debate-washington-university-st-louis-missouri'
url_rep_platform_2016 = 'https://www.presidency.ucsb.edu/documents/2016-republican-party-platform'
url_dem_platform_1980 = 'https://www.presidency.ucsb.edu/documents/1980-democratic-party-platform'
url_rep_platform_1980 = 'https://www.presidency.ucsb.edu/documents/republican-party-platform-1980'
url_debate_cleveland_1980 = 'https://www.presidency.ucsb.edu/documents/presidential-debate-cleveland'
url_debate_baltimore_1980 = 'https://www.presidency.ucsb.edu/documents/presidential-debate-baltimore-reagan-anderson'
```

```python
html_dem_platform_2016 = request.urlopen(url_dem_platform_2016).read().decode('utf8')
raw_dem_platform_2016 = BeautifulSoup(html_dem_platform_2016, 'html.parser')
raw_dem_platform_2016 = [p.get_text() for p in raw_dem_platform_2016.select('div.field-docs-content > p')]
raw_dem_platform_2016 = ' '.join(raw_dem_platform_2016)
tokens_dem_platform_2016 = word_tokenize(raw_dem_platform_2016)
tokens_dem_platform_2016
```

```python
text_dem_platform_2016 = nltk.Text(tokens_dem_platform_2016)
type(text_dem_platform_2016)
```

## Debate Las Vegas 2016

```python
html_debate_las_vegas_2016 = request.urlopen(url_debate_las_vegas_2016).read().decode('utf8')
raw_debate_las_vegas_2016 = BeautifulSoup(html_debate_las_vegas_2016, 'html.parser')

raw_dem_debate_las_vegas_2016 = []
dem_las_vegas_2016 = False
raw_rep_debate_las_vegas_2016 = []
rep_las_vegas_2016 = False

for p in raw_debate_las_vegas_2016.select('div.field-docs-content > p'):
    if p.get_text().startswith('CLINTON'):
        first, _, rest = p.get_text().partition(" ")
        raw_dem_debate_las_vegas_2016.append(rest)
        dem_las_vegas_2016 = True
        rep_las_vegas_2016 = False
    elif p.get_text().startswith('TRUMP'):
        first, _, rest = p.get_text().partition(" ")
        raw_rep_debate_las_vegas_2016.append(rest)
        dem_las_vegas_2016 = False
        rep_las_vegas_2016 = True
    elif p.get_text().startswith('WALLACE'):
        dem_las_vegas_2016 = False
        rep_las_vegas_2016 = False
    elif rep_las_vegas_2016:
        raw_rep_debate_las_vegas_2016.append(p.get_text())
    elif dem_las_vegas_2016:
        raw_dem_debate_las_vegas_2016.append(p.get_text())
        
raw_rep_debate_las_vegas_2016 = ' '.join(raw_rep_debate_las_vegas_2016)
tokens_rep_debate_las_vegas_2016 = word_tokenize(raw_rep_debate_las_vegas_2016)
tokens_rep_debate_las_vegas_2016

raw_dem_debate_las_vegas_2016 = ' '.join(raw_dem_debate_las_vegas_2016)
tokens_dem_debate_las_vegas_2016 = word_tokenize(raw_dem_debate_las_vegas_2016)
tokens_rep_debate_las_vegas_2016
```

## Debate St. Louis 2016

```python
html_debate_st_louis_2016 = request.urlopen(url_debate_st_louis_2016).read().decode('utf8')
raw_debate_st_louis_2016 = BeautifulSoup(html_debate_st_louis_2016, 'html.parser')

raw_dem_debate_st_louis_2016 = []
dem_st_louis_2016 = False
raw_rep_debate_st_louis_2016 = []
rep_st_louis_2016 = False

for p in raw_debate_st_louis_2016.select('div.field-docs-content > p'):
    if p.get_text().startswith('CLINTON'):
        first, _, rest = p.get_text().partition(" ")
        raw_dem_debate_st_louis_2016.append(rest)
        dem_st_louis_2016 = True
        rep_st_louis_2016 = False
    elif p.get_text().startswith('TRUMP'):
        first, _, rest = p.get_text().partition(" ")
        raw_rep_debate_st_louis_2016.append(rest)
        dem_st_louis_2016 = False
        rep_st_louis_2016 = True
    elif p.get_text().startswith('COOPER') or p.get_text().startswith('RADDATZ'):
        dem_st_louis_2016 = False
        rep_st_louis_2016 = False
    elif rep_st_louis_2016:
        raw_rep_debate_st_louis_2016.append(p.get_text())
    elif dem_st_louis_2016:
        raw_dem_debate_st_louis_2016.append(p.get_text())
        
raw_rep_debate_st_louis_2016 = ' '.join(raw_rep_debate_st_louis_2016)
tokens_rep_debate_st_louis_2016 = word_tokenize(raw_rep_debate_st_louis_2016)
tokens_rep_debate_st_louis_2016

raw_dem_debate_st_louis_2016 = ' '.join(raw_dem_debate_st_louis_2016)
tokens_dem_debate_st_louis_2016 = word_tokenize(raw_dem_debate_st_louis_2016)
tokens_dem_debate_st_louis_2016
```

## Debates 2016 combined

```python
tokens_dem_debate_2016 = tokens_dem_debate_las_vegas_2016 + tokens_dem_debate_st_louis_2016
tokens_rep_debate_2016 = tokens_rep_debate_las_vegas_2016 + tokens_rep_debate_st_louis_2016

text_rep_debate_2016 = nltk.Text(tokens_rep_debate_2016)
text_dem_debate_2016 = nltk.Text(tokens_dem_debate_2016)
```

## Debate Cleveland 1980

```python
html_debate_cleveland_1980 = request.urlopen(url_debate_cleveland_1980).read().decode('utf8')
raw_debate_cleveland_1980 = BeautifulSoup(html_debate_cleveland_1980, 'html.parser')

raw_dem_debate_cleveland_1980 = []
dem_cleveland_1980 = False
raw_rep_debate_cleveland_1980 = []
rep_cleveland_1980 = False

for p in raw_debate_cleveland_1980.select('div.field-docs-content > p'):
    if p.get_text().startswith('THE PRESIDENT'):
        first, _, rest = p.get_text().partition(" ")
        first, _, rest = rest.partition(" ")
        raw_dem_debate_cleveland_1980.append(rest)
        dem_cleveland_1980 = True
        rep_cleveland_1980 = False
    elif p.get_text().startswith('GOVERNOR REAGAN'):
        first, _, rest = p.get_text().partition(" ")
        first, _, rest = rest.partition(" ")
        raw_rep_debate_cleveland_1980.append(rest)
        dem_cleveland_1980 = False
        rep_cleveland_1980 = True
    elif p.get_text().upper().startswith('MR. SMITH') or p.get_text().upper().startswith('MR. STONE') or p.get_text().startswith('MR. ELLIS') or p.get_text().startswith('MR. HILLIARD') or p.get_text().startswith('MS. WALTERS'):
        dem_cleveland_1980 = False
        rep_cleveland_1980 = False
    elif rep_cleveland_1980:
        raw_rep_debate_cleveland_1980.append(p.get_text())
    elif dem_cleveland_1980:
        raw_dem_debate_cleveland_1980.append(p.get_text())
        
raw_rep_debate_cleveland_1980 = ' '.join(raw_rep_debate_cleveland_1980)
tokens_rep_debate_cleveland_1980 = word_tokenize(raw_rep_debate_cleveland_1980)
tokens_rep_debate_cleveland_1980

raw_dem_debate_cleveland_1980 = ' '.join(raw_dem_debate_cleveland_1980)
tokens_dem_debate_cleveland_1980 = word_tokenize(raw_dem_debate_cleveland_1980)
tokens_dem_debate_cleveland_1980
tokens_rep_debate_cleveland_1980
```

## Debate Baltimore 1980

```python
html_debate_baltimore_1980 = request.urlopen(url_debate_baltimore_1980).read().decode('utf8')
raw_debate_baltimore_1980 = BeautifulSoup(html_debate_baltimore_1980, 'html.parser')

raw_dem_debate_baltimore_1980 = []
dem_baltimore_1980 = False
raw_rep_debate_baltimore_1980 = []
rep_baltimore_1980 = False

for p in raw_debate_baltimore_1980.select('div.field-docs-content > p'):
    if p.get_text().startswith('REP. JOHN B. ANDERSON') or p.get_text().startswith('ANDERSON'):
        first, _, rest = p.get_text().partition(" ")
        raw_dem_debate_baltimore_1980.append(rest)
        dem_baltimore_1980 = True
        rep_baltimore_1980 = False
    elif p.get_text().startswith('GOV. RONALD REAGAN') or p.get_text().startswith('REAGAN'):
        first, _, rest = p.get_text().partition(" ")
        raw_rep_debate_baltimore_1980.append(rest)
        dem_baltimore_1980 = False
        rep_baltimore_1980 = True
    elif p.get_text().upper().startswith('MOYERS') or p.get_text().upper().startswith('LOOMIS') or p.get_text().startswith('GREENBERG') or p.get_text().startswith('CHARLES CORDDRY') or p.get_text().startswith('CORDDRY') or p.get_text().startswith('LEE MAY') or p.get_text().startswith('MAY') or p.get_text().startswith('JANE BRYANT QUINN') or p.get_text().startswith('QUINN') or p.get_text().startswith('GOLDEN'):
        dem_baltimore_1980 = False
        rep_baltimore_1980 = False
    elif rep_baltimore_1980:
        raw_rep_debate_baltimore_1980.append(p.get_text())
    elif dem_baltimore_1980:
        raw_dem_debate_baltimore_1980.append(p.get_text())
        
raw_rep_debate_baltimore_1980 = ' '.join(raw_rep_debate_baltimore_1980)
tokens_rep_debate_baltimore_1980 = word_tokenize(raw_rep_debate_baltimore_1980)
tokens_rep_debate_baltimore_1980

raw_dem_debate_baltimore_1980 = ' '.join(raw_dem_debate_baltimore_1980)
tokens_dem_debate_baltimore_1980 = word_tokenize(raw_dem_debate_baltimore_1980)
tokens_dem_debate_baltimore_1980
```
