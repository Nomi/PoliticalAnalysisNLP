```python
from bs4 import BeautifulSoup
from nltk import word_tokenize
from urllib import request
import nltk
```

## Democratic Party

```python
url_dem_platform = 'https://www.presidency.ucsb.edu/documents/2016-democratic-party-platform'
url_rep_platform = 'https://www.presidency.ucsb.edu/documents/2016-republican-party-platform'
```

```python
html_dem_platform = request.urlopen(url_dem_platform).read().decode('utf8')
raw_dem_platform = BeautifulSoup(html_dem_platform, 'html.parser')
raw_dem_platform = [p.get_text() for p in raw_dem_platform.select('div.field-docs-content > p')]
raw_dem_platform = ' '.join(raw_dem_platform)
tokens_dem_platform = word_tokenize(raw_dem_platform)
tokens_dem_platform
```

```python
text_dem_platform = nltk.Text(tokens_dem_platform)
type(text_dem_platform)
```

## Republican Party

```python
html_rep_platform = request.urlopen(url_rep_platform).read().decode('utf8')
raw_rep_platform = BeautifulSoup(html_rep_platform, 'html.parser')
raw_rep_platform = [p.get_text() for p in raw_rep_platform.select('div.field-docs-content > p')]
raw_rep_platform = ' '.join(raw_rep_platform)
tokens_rep_platform = word_tokenize(raw_rep_platform)
tokens_rep_platform
```

```python
text_rep_platform = nltk.Text(tokens_rep_platform)
type(text_rep_platform)
```
